#### Check Missing Values

```python
na_columns = data.columns[data.isna().any()].tolist()
for clmn in na_columns:
    print(data[clmn].dtypes)
    print(data[clmn].unique())
```

#### Log Normalization

```python
y = data.SalePrice
print("Skewness: %f" % y.skew())
print("Kurtosis: %f" % y.kurt())
print("Min: %f, Max: %f" % (y.min(), y.max()))
```

```python
sns.displot(y)
```

```python
# Only applicable for right-skewed data
log_y = np.log(y)
print("Skewness: %f" % log_y.skew())
print("Kurtosis: %f" % log_y.kurt())
print("Min: %f, Max: %f" % (log_y.min(), log_y.max()))
```

```python
sns.displot(log_y)
```
